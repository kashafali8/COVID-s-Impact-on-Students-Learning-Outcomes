```python
# get a list of ENTITY_CD that are in boces_2019 but not in boces_2022
code_2019 = boces_2019['ENTITY_CD'].unique().tolist()

# get a list of ENTITY_CD that are in boces_2022 but not in boces_2019
code_2022 = boces_2022['ENTITY_CD'].unique().tolist()

# check the difference between the two lists
print(f"The difference between the two lists is {set(code_2019).difference(set(code_2022))}")
```

```python
# select the year to be 2018 and 2019 for boces_2019
boces_2019 = boces_2019[boces_2019['YEAR'].isin([2018,2019])]

lst_2018 = boces_2019[boces_2019['YEAR']==2018]['ENTITY_CD'].tolist()
lst_2019 = boces_2019[boces_2019['YEAR']==2019]['ENTITY_CD'].tolist()

# select the year to be 2020 and 2021 for boces_2022
boces_2022 = boces_2022[boces_2022['YEAR'].isin([2021,2022])]

lst_2021 = boces_2022[boces_2022['YEAR']==2021]['ENTITY_CD'].tolist()
lst_2022 = boces_2022[boces_2022['YEAR']==2022]['ENTITY_CD'].tolist()
```

```python
boces_2018 = boces_2019[boces_2019['YEAR']==2018]
boces_2019 = boces_2019[boces_2019['YEAR']==2019]
boces_2021 = boces_2022[boces_2022['YEAR']==2021]
boces_2022 = boces_2022[boces_2022['YEAR']==2022]

# make a dataframe to stack the data by concatenating the dataframes
boces_2018['YEAR'] = 2018
boces_2019['YEAR'] = 2019
boces_2021['YEAR'] = 2021
boces_2022['YEAR'] = 2022

# find out the overlap ENTITY_CD across four years using lst_2018, lst_2019, lst_2021, lst_2022
#print(f"The overlap ENTITY_CD across four years is {set(lst_2018).intersection(set(lst_2019)).intersection(set(lst_2021)).intersection(set(lst_2022))}")

entity_id_four = list(set(lst_2018).intersection(set(lst_2019)).intersection(set(lst_2021)).intersection(set(lst_2022)))
len(entity_id_four)
```

```python
# concat the dataframes of boces_2022 and boces_2019
boces_2018_2019_2021_2022 = pd.concat([boces_2018, boces_2019,boces_2022, boces_2021],axis=0)
# get the unique ENTITY_CD count
boces_2018_2019_2021_2022['ENTITY_CD'].nunique()
# total_length = len(lst_2018) + len(lst_2019) + len(lst_2021) + len(lst_2022)
```

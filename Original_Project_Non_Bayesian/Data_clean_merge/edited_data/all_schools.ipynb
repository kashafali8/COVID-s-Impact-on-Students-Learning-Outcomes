## Merge 2018-2019 and 2021-2022

```python
# concat all the dataframes

df_total = pd.concat(
    [
        df_ela_2018_19_total,
        df_math_2018_19_total,
        df_sci_2018_19_total,
        df_ela_2021_22_total,
        df_math_2021_22_total,
        df_sci_2021_22_total,
    ],
    ignore_index=True,
)
```

```python
df_total.ASSESSMENT_NAME.unique()
```

```python
df_total.dtypes
```

```python
df_total["ES_ELA"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "ELA3",
                "ELA4",
                "ELA5"
       
            ]
        ).intersection(set(x)) != set())
    )
```

```python
df_total["ES_MATH"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "MATH3",
                "MATH4",
                "MATH5"
       
            ]
        ).intersection(set(x)) != set())
    )
```

```python
df_total["ES_SCI"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "Science4"       
            ]
        ).intersection(set(x)) != set())
)
```

```python
df_total.head(20)
```

```python
# cases where ES_ELA is 1 but ES_MATH is 0 and ES_SCI is 0

df_total[
    (df_total["ES_ELA"] == 1)
    & (df_total["ES_MATH"] == 1)
    & (df_total["ES_SCI"] == 0)
    
]
```

```python
df_total[(df_total["ENTITY_CD"] == 320800861074) & (df_total["YEAR"] == 2019)]
```

```python
# if any one of ES_ELA, ES_MATH, ES_SCI is 1, then ES is 1

df_total["ES"] = df_total["ES_ELA"] | df_total["ES_MATH"] | df_total["ES_SCI"]
```

```python
df_total["MS_ELA"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "ELA6",
                "ELA7",
                "ELA8"
       
            ]
        ).intersection(set(x)) != set())
    )
```

```python
df_total["MS_MATH"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "MATH6",
                "MATH7",
                "MATH8"
       
            ]
        ).intersection(set(x)) != set())
    )
```

```python
df_total["MS_SCI"] = df_total.groupby(["ENTITY_CD", "YEAR"])[
    "ASSESSMENT_NAME"
].transform(
    lambda x: int(
        set(
            [
                "Science8"       
            ]
        ).intersection(set(x)) != set())
)
```

```python
# if any one of MS_ELA, MS_MATH, MS_SCI is 1, then MS is 1

df_total["MS"] = df_total["MS_ELA"] | df_total["MS_MATH"] | df_total["MS_SCI"]
```

```python
df_total.head()
```

```python
# both is 1 if both ES and MS are 1

df_total["EM"] = df_total["ES"] & df_total["MS"]

# st_notsure is 1 if both ES and MS are 0
subset = df_total[(df_total["ES"] == 0) & (df_total["MS"] == 0)]
df_total.loc[subset.index, "ST_NOTSURE"] = 1
df_total["ST_NOTSURE"] = df_total["ST_NOTSURE"].fillna(0)

df_total.head()
```

```python
# save the dataframe to a csv file

df_total.to_csv("all_students.csv" , index=False)
```

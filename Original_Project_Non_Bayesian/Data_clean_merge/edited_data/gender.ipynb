## ELA 2018-2019 Gender

```python
df_ela_2018_19_total = pd.read_csv(
    "/Users/kashafali/Documents/Duke/Spring23/UDS/unifying-data-science-2023-project-grupo9/Data_clean_merge/edited_data/ELA_edited_2018_19.csv"
)

df_ela_2018_19_total.head()
```

```python
df_ela_2018_19_total.isnull().sum()
```

```python
df_ela_2018_19_gender = df_ela_2018_19_total[
    (df_ela_2018_19_total["SUBGROUP_NAME"] == "Female")
    | (df_ela_2018_19_total["SUBGROUP_NAME"] == "Male")
]
```

```python
# fix index
df_ela_2018_19_gender = df_ela_2018_19_gender.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_ela_2018_19_gender = df_ela_2018_19_gender.reindex(
    columns=sorted(df_ela_2018_19_gender.columns, key=lambda x: x[::-1])
)

df_ela_2018_19_gender.columns = [
    "{}_{}".format(t, v) for v, t in df_ela_2018_19_gender.columns
]

df_ela_2018_19_gender = pd.DataFrame(df_ela_2018_19_gender.to_records())
```

```python
df_ela_2018_19_gender.head()
```

```python
df_ela_2018_19_gender.columns

# drop Female_ENTITY_NAME
df_ela_2018_19_gender = df_ela_2018_19_gender.drop(
    ["Female_ENTITY_NAME", "Male_ENTITY_NAME"], axis=1
)
```

```python
print("df_ela_2018_19_gender.shape: ", df_ela_2018_19_gender.shape)

df_ela_2018_19_gender.isnull().sum()
```

```python

def sanity_checking_col_names(df_list, list_names, key_col):
    """
    This function is used to check the column names of the dataframes
    """

    unique_set = set()

    for name in list_names:

        # print(name)

        for df in df_list:

            assert "SUBGROUP_NAME" in df.columns

            for val in df[key_col].unique():

                if name in val.lower():

                    unique_set.add(val)

    return unique_set
```

```python
def establish_composition(main_df, df_list, base_cols, subgroups):
    """
    This function is used to establish the composition of the subgroups
    """''
    core_df = main_df.copy() # to contain all the core records

    # copy_df['Subject'] = copy_df['ASSESSMENT_NAME'].str.extract(r"([A-Za-z]*).*", expand=False)

    # copy_df[copy_df.MEAN_SCORE == 's'] = np.nan

    # copy_df['MEAN_SCORE'] = copy_df['MEAN_SCORE'].astype(float)

    # copy_df["MEAN_SCORE"].fillna(copy_df.groupby(["YEAR", "ENTITY_CD", "Subject"])["MEAN_SCORE"].transform("mean"), inplace=True)

    composition_df = pd.DataFrame(columns = base_cols + subgroups)

    for df in df_list:
        
        temp_df = df.copy()

        temp_df['Subject'] = temp_df['ASSESSMENT_NAME'].str.extract(r"([A-Za-z]*).*", expand=False)

        redux_df = temp_df[base_cols].copy()
        redux_df.drop_duplicates(inplace=True, ignore_index=True) # remove duplicates

        for group in subgroups:

            group_df = temp_df[temp_df["SUBGROUP_NAME"] == group].copy()

            # print(group_df['NUM_TESTED'].isna().sum(), group_df['NOT_TESTED'].isna().sum())
            try: 
                group_df['NUM_TESTED'] = group_df['NUM_TESTED'].astype(float)
                group_df['NOT_TESTED'] = group_df['NOT_TESTED'].astype(float)
            except:
                group_df['NUM_TESTED'] = group_df['NUM_TESTED'].replace('s', np.nan)
                group_df['NOT_TESTED'] = group_df['NOT_TESTED'].replace('s', np.nan)
                group_df['NUM_TESTED'] = group_df['NUM_TESTED'].astype(float)
                group_df['NOT_TESTED'] = group_df['NOT_TESTED'].astype(float)

            group_df[f'{group}'] = group_df['NUM_TESTED'] + group_df['NOT_TESTED']

            group_df = group_df[base_cols + [f'{group}']].copy()

            redux_df = redux_df.merge(group_df, on=base_cols, how='left')

        composition_df = pd.concat([composition_df, redux_df], ignore_index=True)

    composition_df.drop_duplicates(subset=base_cols, inplace=True, ignore_index=True)

    core_df = core_df.merge(composition_df, on=base_cols, how='left')

    core_df.fillna(0, inplace=True)

    core_df['female_%'] = core_df['Female'] / core_df['All Students']
    core_df['male_%'] = core_df['Male'] / core_df['All Students']

    return core_df
            


    #     new_df_list.append(df[base_cols + subgroups].copy())




    
    # if keep_cols != []:
    #     new_cols = keep_cols + base_cols + ["SUBGROUP_NAME"]

    # else:

    #     new_cols = base_cols + ["SUBGROUP_NAME"]

    # base_df = copy_df[new_cols].copy() # get the core records

    # base_df.drop_duplicates(subset = base_cols, inplace=True, ignore_index=True) # remove duplicates

    # core_df = copy_df[new_cols].copy() # get the core records

    # # At this point, we have a df custom to merge with Team 9's base data
    # # that contains all the schools,
    # # But we still need the data by race

    # # We will create a dictionary of dataframes

    # for group in subgroups:

    #     # copy to avoid pointer issues
        
    #     group_df = core_df[core_df["SUBGROUP_NAME"] == group].copy()

    #     formatted_name = group.replace(" ", "_").lower()

    #     base_df = base_df.merge(group_df, on=base_cols, how="left", suffixes=("", f"_{formatted_name}"))

    # # Removing the subgroup name columns

    # ls_subgroup_cols = [col for col in base_df.columns if "subgroup_name" in col.lower()]

    # extra_cols = ['NUM_TESTED', 'NUM_PROF', 'TOTAL_SCALE_SCORES', 'MEAN_SCORE']

    # base_df.drop(columns=ls_subgroup_cols + extra_cols, inplace=True)
    
    # return base_df
    
```

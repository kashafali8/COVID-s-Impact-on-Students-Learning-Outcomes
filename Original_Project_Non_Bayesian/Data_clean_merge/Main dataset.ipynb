```python
import pandas as pd
import numpy as np
```

```python
subject_data = pd.read_csv('subject_df_2018_19_2021_22.csv')
```

```python
subject_data.head()
```

```python
df_2018 = subject_data[subject_data['YEAR'] == 2018]
```

```python
df_2018 = df_2018[['ENTITY_CD', 'ENTITY_NAME', 'YEAR', 'ASSESSMENT_NAME', 'NOT_TESTED', 'NUM_TESTED', 'NUM_PROF', 'PER_PROF', 'TOTAL_SCALE_SCORES', 'MEAN_SCORE', 'SUBJECT']]
```

```python
# count rows
df_2018.shape
```

```python
df_2018.head()
```

```python
df_2018[df_2018["ENTITY_CD"] == 1]
```

```python
df_eng = df_2018[df_2018["SUBJECT"] == "ELA"]
df_eng.head(7)
```

```python
df_eng.dtypes
```

```python
# identify any character in total_scale_scores
df_eng[df_eng['TOTAL_SCALE_SCORES'].str.contains(r'[a-zA-Z]')]

# replace characters with nan
df_eng['TOTAL_SCALE_SCORES'] = pd.to_numeric(df_eng['TOTAL_SCALE_SCORES'].str.extract('(\d+)', expand=False), errors='coerce')

df_eng['MEAN_SCORE'] = pd.to_numeric(df_eng['MEAN_SCORE'].str.extract('(\d+)', expand=False), errors='coerce')
```

```python
# convert num_tested to float
df_eng['NUM_TESTED'] = df_eng['NUM_TESTED'].astype(float)
```

```python
df_eng.dtypes
```

```python
df_eng.head()
```

```python
df_eng['TOTAL_SCALE_ELA'] = df_eng.groupby(['ENTITY_CD'])['TOTAL_SCALE_SCORES'].transform('sum')
df_eng['MEAN_ELA'] = np.round(df_eng['TOTAL_SCALE_ELA'] / df_eng['NUM_TESTED'])
```

```python
df_eng.head()
```

```python
# replace total_scale_scores with total_scale_ela where Assessment_name = 'ELA3_8' if total_scale_scores = 0
df_eng.loc[((df_eng['ASSESSMENT_NAME'] == 'ELA3_8') & (df_eng['TOTAL_SCALE_SCORES'] == 00)), 'TOTAL_SCALE_SCORES'] = df_eng['TOTAL_SCALE_ELA']
```

```python
df_eng.head()
```

```python
# replace total_scale_scores with total_scale_ela where Assessment_name = 'ELA3_8' if total_scale_scores = 0
df_eng.loc[((df_eng['ASSESSMENT_NAME'] == 'ELA3_8') & (df_eng['MEAN_SCORE'] == 00)), 'MEAN_SCORE'] = df_eng['MEAN_ELA']
```

```python
# subset df_eng to only include rows where Assessment_name = 'ELA3_8'
df_eng_2018 = df_eng[df_eng['ASSESSMENT_NAME'] == 'ELA3_8']
```

```python
def merge_to_base(base_df : pd.DataFrame, df_dict : dict, keep_cols : list[str] = None) -> pd.DataFrame:
    """Takes a base dataframe and a dictionary of dataframes
    and returns a dataframe with all the data merged
    """
    for key in df_dict.keys():

        combined_df = base_df.merge(df_dict[key], on=["ENTITY_CD", "YEAR", "Subject", "ASSESSMENT_NAME"], how="inner")

    combined_df_col_list = [col for col in combined_df.columns]

    if keep_cols is (not None):

        keep_list = []

        for col in keep_cols:

            if col in combined_df.columns:

                keep_list.append(col)

        combined_df = (combined_df[["ENTITY_CD", "YEAR", "Subject", "ASSESSMENT_NAME"] + keep_list]).copy()

    return combined_df
```

```python
def df_stacker(full_df : pd.DataFrame, to_stack_df : pd.DataFrame) -> pd.DataFrame:
    """Takes a dataframe and stacks it on the full_df
    """
    # Get the columns that are not in the full_df
    
    stacked_df =  pd.concat([full_df, to_stack_df], ignore_index = True, axis=0)

    return stacked_df
```

```python
dfs = []
for i in df_path:
    if "_Annual" in i:
        df = pd.read_csv(os.path.join("/Users/belladu/Desktop/IDS701 Unifying Data Science/unifying-data-science-2023-project-grupo9-main 2/00_source_data/00_source_data", i))
        df = df[['ENTITY_CD', 'YEAR', 'ASSESSMENT_NAME']]
        df['Subject'] = i.split(" ")[2].replace(".csv", "")
        dfs.append(df)
```

```python
df = pd.DataFrame()
for i in dfs:
    df = pd.concat([df, i], axis=0)
del dfs
```

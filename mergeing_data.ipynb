```python
# read science and math 2019 and 2022 data
df_science_2019 = pd.read_csv(scienc_2019_path)
df_science_2022 = pd.read_csv(scienc_2022_path)
df_math_2019 = pd.read_csv(math_2019_path)
df_math_2022 = pd.read_csv(math_2022_path)
df_ela_2019 = pd.read_csv(ela_2019_path)
df_ela_2022 = pd.read_csv(ela_2022_path)
df_boces = pd.read_csv(boces_path)

df_science_2019.head()
```

```python
# get the unique EntityID in 2019
df_science_2019_unique = df_science_2019['ENTITY_CD'].unique()
df_science_2019_unique
```

```python
# add a subject label to the 2019 data
df_science_2019['Subject'] = 'Science'
df_math_2019['Subject'] = 'Math'
df_ela_2019['Subject'] = 'ELA'

# add a year label to the 2019 data
df_science_2019['YEAR'] = 2019
df_math_2019['YEAR'] = 2019
df_ela_2019['YEAR'] = 2019

# add a subject label to the 2022 data
df_science_2022['Subject'] = 'Science'
df_math_2022['Subject'] = 'Math'
df_ela_2022['Subject'] = 'ELA'

# add a year label to the 2022 data
df_science_2022['YEAR'] = 2022
df_math_2022['YEAR'] = 2022
df_ela_2022['YEAR'] = 2022

```

```python
# combine the 2019 and 2022 data
df_science = pd.concat([df_science_2019, df_science_2022], axis=0)
df_math = pd.concat([df_math_2019, df_math_2022], axis=0)
df_ela = pd.concat([df_ela_2019, df_ela_2022], axis=0)

# combine the science, math, and ela data
df_subject = pd.concat([df_science, df_math, df_ela], axis=0)

# get the unique EntityID in 2019
df_subject_unique = df_subject['ENTITY_CD'].unique()
df_subject_unique
```
